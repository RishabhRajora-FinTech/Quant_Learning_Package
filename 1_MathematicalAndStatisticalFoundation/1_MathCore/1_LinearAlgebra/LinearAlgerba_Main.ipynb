## Key Formulas

**Dot Product:**

$$\vec{a} \cdot \vec{b} = \sum_{i=1}^{n} a_i b_i = \|a\| \|b\| \cos\theta$$

**Matrix Multiplication:**

$$C_{ij} = \sum_{k=1}^{n} A_{ik} B_{kj}$$

**Eigenvalue Equation:**

$$A\vec{v} = \lambda\vec{v}$$

**SVD Decomposition:**

$$A = U \Sigma V^T$$
